# Data Cleaning and Aggregation

Covers selection, filtering, sorting, missing-value handling, column derivation, renaming, group aggregation, merging, and applying custom functions on a small sample dataset.

```python
import pandas as pd
```

## Create Sample Dataset

```python
data = {
    'Name': ['A1', 'A2', 'A3', 'A4', 'A5'],
    'Age': [20, None, 21, None, 27],
    'Grade': ['A', 'C', 'B', 'D', 'B'],
    'City': ['XY', None, 'YZ', 'YY', None],
}

df = pd.DataFrame(data)
df
```

## Handling Missing Values

```python
df.info()
```

```python
df.isnull()
```

```python
# Fill missing Age values with the column mean
df['Age'] = df['Age'].fillna(df['Age'].mean())

# Drop rows where City is missing
df = df.dropna(subset=['City'])
df
```

## Adding a Derived Column

```python
df['Age*2'] = df['Age'] * 2
df
```

## Renaming Columns

```python
df = df.rename(columns={'Name': 'Full_Name'})
df
```

## Group By and Aggregation

```python
agg_df = df.groupby('Grade').agg({'Age': 'mean', 'Age*2': 'sum'})
agg_df
```

## Merging with Additional Data

```python
extra_data = pd.DataFrame({
    'Full_Name': ['A1', 'A3', 'A5'],
    'Score': [88, 92, 79],
})

merged_df = pd.merge(df, extra_data, on='Full_Name', how='left')
merged_df
```

## Applying a Custom Function

```python
df['Age_Category'] = df['Age'].apply(lambda x: 'Young' if x < 25 else 'Adult')
df
```
