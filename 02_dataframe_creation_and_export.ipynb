# DataFrame Creation and Export

Builds a small DataFrame from scratch, inspects descriptive statistics, and exports it to multiple file formats (Excel, JSON, CSV).

```python
import pandas as pd
```

## Create DataFrame

```python
data = {
    'Name': ['Hamid Raza', 'Ali', 'Hassan', 'Hussain'],
    'Age': [21, 18, 16, 12],
    'Class': ['BSCS', 'Intermediate', 'First Year', 'O-level'],
}

df = pd.DataFrame(data)
df
```

## Descriptive Statistics

```python
df.describe()
```

## Export to Excel, JSON, and CSV

```python
OUTPUT_DIR = '../outputs'

df.to_excel(f'{OUTPUT_DIR}/students.xlsx', index=False)
df.to_json(f'{OUTPUT_DIR}/students.json', indent=2)
df.to_csv(f'{OUTPUT_DIR}/students.csv', index=False)
print('Exported to outputs/students.xlsx, .json, .csv')
```

## Verify Exported Files

```python
pd.read_excel(f'{OUTPUT_DIR}/students.xlsx')
```

```python
pd.read_json(f'{OUTPUT_DIR}/students.json')
```

```python
pd.read_csv(f'{OUTPUT_DIR}/students.csv')
```
