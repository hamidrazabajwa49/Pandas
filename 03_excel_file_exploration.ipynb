# Excel File Exploration

Loads an external Excel dataset and inspects its structure: shape, columns, size, and summary info.

**Data source**: `../data/practicedata1.xlsx`

```python
import pandas as pd

DATA_PATH = 'data/practicedata.xlsx'

df = pd.read_excel(DATA_PATH)
df.head()
```

## Tail Rows

```python
df.tail()
```

## Shape, Columns, Size

```python
print('Shape:', df.shape)
print('Columns:', df.columns.tolist())
print('Size:', df.size)
```

## Data Types and Non-Null Counts

```python
df.info()
```
