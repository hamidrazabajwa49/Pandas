# Pandas Basics Reference

Quick reference covering core pandas operations: creation, inspection, selection, filtering, modification, sorting, missing values, aggregation, grouping, merging, concatenation, indexing, and I/O.

```python
import pandas as pd
```

## 1. Creating a DataFrame and Series

```python
df = pd.DataFrame({'Name': ['Ali', 'Sara', 'Ahmed'], 'Age': [22, 25, 24]})
df
```

## 4. Filtering Rows

```python
df[df['Age'] > 23]
```

```python
df[(df['Age'] > 22) & (df['Age'] < 25)]
```

## 5. Modifying Data

```python
df['Age'] = df['Age'] + 1
df['AgeDoubled'] = df['Age'] * 2
df
```

```python
df = df.rename(columns={'Age': 'Years'})
df = df.drop(columns='AgeDoubled')
df
```

## 6. Sorting

```python
df.sort_values(by='Years')
```

```python
df.sort_values(by='Years', ascending=False)
```

## 7. Handling Missing Values

```python
df_nan = df.copy()
df_nan.loc[1, 'Years'] = None
df_nan.isnull()
```

```python
df_nan.dropna()
```

```python
df_nan.fillna(0)
```

## 8. Aggregation

```python
print('Sum:', df['Years'].sum())
print('Mean:', df['Years'].mean())
print('Max:', df['Years'].max())
print('Min:', df['Years'].min())
print('Count:', df['Years'].count())
```

## 9. Group By

```python
df.groupby('Years')['Name'].count()
```

## 10. Apply Function

```python
df['Years'].apply(lambda x: x * 2)
```

## 11. Merging

```python
df1 = pd.DataFrame({'ID': [1, 2], 'Name': ['Ali', 'Sara']})
df2 = pd.DataFrame({'ID': [1, 2], 'Score': [90, 85]})

pd.merge(df1, df2, on='ID')
```

## 12. Concatenation

```python
pd.concat([df1, df2], axis=1)
```

## 13. Reset and Set Index

```python
indexed_df = df.set_index('Name')
indexed_df
```

```python
indexed_df.reset_index()
```

```python
OUTPUT_PATH = '../outputs/basics_reference.csv'

df.to_csv(OUTPUT_PATH, index=False)
pd.read_csv(OUTPUT_PATH)
```
