# Advanced Pandas Techniques

Covers duplicate handling, value replacement, string operations, dtype conversion, multi-condition filtering, `query`, pivot tables, multi-index DataFrames, and exporting cleaned data.

```python
import pandas as pd
```

## Setup: Recreate Cleaned Dataset

Same end state as `05_data_cleaning_and_aggregation.ipynb`.

```python
data = {
    'Full_Name': ['A1', 'A2', 'A3', 'A4', 'A5'],
    'Age': [20.0, None, 21.0, None, 27.0],
    'Grade': ['A', 'C', 'B', 'D', 'B'],
    'City': ['XY', None, 'YZ', 'YY', None],
}

df = pd.DataFrame(data)
df['Age'] = df['Age'].fillna(df['Age'].mean())
df = df.dropna(subset=['City'])
df['Age*2'] = df['Age'] * 2
df['Age_Category'] = df['Age'].apply(lambda x: 'Young' if x < 25 else 'Adult')
df
```

## Detecting Duplicates

```python
dup_df = pd.DataFrame({
    'ID': [1, 2, 2, 3, 4, 4],
    'Value': ['X', 'Y', 'Y', 'Z', 'X', 'X'],
})
dup_df.duplicated()
```

## Dropping Duplicates

```python
dup_df = dup_df.drop_duplicates()
dup_df
```

## Replacing Values

```python
grade_labels = {'A': 'Excellent', 'B': 'Good', 'C': 'Average', 'D': 'Poor'}
df['Grade'] = df['Grade'].replace(grade_labels)
df
```

## String Operations

```python
df['Full_Name'] = df['Full_Name'].str.lower()
df
```

## Changing Data Types

```python
df['Age'] = df['Age'].astype(float)
df.dtypes
```

## Multi-Condition Filtering

```python
df[(df['Age'] > 20) & (df['Grade'] == 'Good')]
```

## Filtering with `query`

```python
df.query("Age >= 21 and Grade == 'Good'")
```

## Pivot Table

```python
pivot = df.pivot_table(values='Age', index='Grade', columns='City', aggfunc='mean')
pivot
```

## Multi-Index DataFrame

```python
multi_idx_df = df.set_index(['City', 'Grade'])
multi_idx_df
```

## Export Cleaned Data

```python
OUTPUT_PATH = '../outputs/cleaned_data.csv'

df.to_csv(OUTPUT_PATH, index=False)
print(f'Cleaned data saved to {OUTPUT_PATH}')
```
