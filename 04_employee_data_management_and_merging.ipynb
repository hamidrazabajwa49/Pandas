# Employee Data Management and Merging

Covers column updates, missing-value handling, sorting, grouped aggregation, and the four merge types (inner, left, right, outer).

```python
import pandas as pd
```

## Create Employee Dataset

```python
data = {
    'ID': [101, 102, 103, 104, 105],
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Department': ['HR', 'IT', 'Finance', 'IT', 'HR'],
    'Joining_Date': pd.to_datetime(
        ['2020-01-15', '2019-03-10', '2021-06-25', '2018-07-30', '2022-11-05']
    ),
    'Salary': [50000, 60000, 55000, 70000, 48000],
    'Performance_Score': [88, 92, 79, 95, 85],
}

df = pd.DataFrame(data)
df
```

## Update an Existing Column

Apply a 10% salary increase.

```python
df['Salary'] = df['Salary'] * 1.10
df
```

## Handling Missing Data

```python
df_missing = df.copy()
df_missing.loc[2, 'Salary'] = None
df_missing.isnull()
```

```python
# Fill missing salary with the column mean
df_filled = df_missing.copy()
df_filled['Salary'] = df_filled['Salary'].fillna(df_filled['Salary'].mean())
df_filled
```

```python
# Alternative: fill via linear interpolation
df_interpolated = df_missing.copy()
df_interpolated['Salary'] = df_interpolated['Salary'].interpolate(method='linear')
df_interpolated
```

## Sorting

```python
df.sort_values(by='Performance_Score', ascending=True)
```

```python
df.sort_values(by=['Performance_Score', 'ID'], ascending=[True, True])
```

## Grouping and Aggregation

```python
df.groupby('Department')['Salary'].sum()
```

## Merging DataFrames

Two related tables: employee records and department metadata.

```python
employees_df = pd.DataFrame({
    'employee_id': [101, 102, 103, 104, 105, 106],
    'name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve', 'Frank'],
    'department_id': [10, 20, 10, 30, 20, 40],
    'hire_date': ['2022-01-15', '2021-05-20', '2023-03-10',
                   '2022-11-01', '2024-02-28', '2023-08-12'],
})

departments_df = pd.DataFrame({
    'department_id': [10, 20, 30, 50],
    'department_name': ['Marketing', 'Sales', 'Engineering', 'HR'],
    'manager': ['Sarah', 'Mike', 'Jessica', 'Kevin'],
})

employees_df
```

```python
departments_df
```

### Inner Merge

Only rows with a matching `department_id` in both tables. Frank (dept 40) and HR (dept 50) are dropped.

```python
pd.merge(employees_df, departments_df, on='department_id', how='inner')
```

### Left Merge

All employees retained; unmatched department info becomes NaN (Frank).

```python
pd.merge(employees_df, departments_df, on='department_id', how='left')
```

### Right Merge

All departments retained; unmatched employee info becomes NaN (HR).

```python
pd.merge(employees_df, departments_df, on='department_id', how='right')
```

### Outer Merge

All rows from both tables, with NaN where there is no match.

```python
pd.merge(employees_df, departments_df, on='department_id', how='outer')
```
